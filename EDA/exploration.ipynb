### Explore the dataset

```python
from dotenv import load_dotenv
import os
import kagglehub
import pandas as pd
import numpy as np

load_dotenv()
```

```python
# Download the PCB defects dataset to our data directory
import shutil
from pathlib import Path

data_dir = Path("/Users/bahacelik/Documents/Coding/Automated-Quality-Inspection-Station/data")

# Download dataset (goes to kagglehub cache first)
cache_path = kagglehub.dataset_download("akhatova/pcb-defects")
print(f"Downloaded to cache: {cache_path}")

# Copy to our data directory
if data_dir.exists() and any(data_dir.iterdir()):
    print(f"Data directory already has contents, skipping copy")
    dataset_path = data_dir
else:
    # Copy all contents from cache to data directory
    for item in Path(cache_path).iterdir():
        dest = data_dir / item.name
        if item.is_dir():
            shutil.copytree(item, dest, dirs_exist_ok=True)
        else:
            shutil.copy2(item, dest)
    print(f"Copied dataset to: {data_dir}")
    dataset_path = data_dir

print(f"\nDataset path: {dataset_path}")
```
